```python
from Bio import SeqIO
import json

g1 = "../data/hg38/hg38.ml.fa"
g1_records = {}
for record in SeqIO.parse(g1, "fasta"):
    g1_records[record.id] = record.seq 
```

```python
import json
g2_assembly = "../data/hg38/ncbi_dataset/data/GCF_000001405.26/sequence_report.jsonl"
data = []
with open(g2_assembly, "r") as f:
    for line in f:
        # Avoid empty lines
        if line.strip():  
            data.append(json.loads(line))
```

```python
g2_records_names = {}
for assembly in data:
    if assembly["ucscStyleName"] in g1_records:
        g2_records_names[assembly["refseqAccession"]] = assembly["ucscStyleName"]
```

```python
import re

g2 = "../data/hg38/ncbi_dataset/data/GCF_000001405.26/GCF_000001405.26_GRCh38_genomic.fna"
for record in SeqIO.parse(g2, "fasta"):
    if record.id in g2_records_names:
        g1_seq = str(g1_records[g2_records_names[record.id]]).lower()
        g2_seq = str(record.seq).lower()
        g2_seq = re.sub(r"[^atcgn]", "n", g2_seq)
        print(g1_seq == g2_seq)
        
        #for i, (c1, c2) in enumerate(zip(str(g1_seq).lower(), str(g2_seq).lower())):
        #    if c1 != c2:
        #        print(i)
```
